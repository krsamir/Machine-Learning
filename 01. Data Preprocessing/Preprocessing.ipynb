```python
from sklearn.preprocessing import Imputer    #Importing Library
#Imputer will take care of missing Datas
```

**Categorical Data**


Here we can see that there are 2 datas Which can be categorised that is Country and Purchased.
Where Country contains 3 categories(France,Spain,Germany) and Purchased contain 2 categories(Yes,no).

We have to encode the text into numbers so that it is understandable by the machine

```python
#Encoding Categorical Data
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder_X = LabelEncoder()    #Calling Label Encoder Class
```

```python
labelencoder_X.fit_transform(X[:,0])
```

```python
X
```

```python
X[:,0] = labelencoder_X.fit_transform(X[:,0])#fit transform function transform and fits data.
```

```python
X
```

We will be using dummy variables i.e. making the same number of columns as no of categories. This is done because if the country are numbered 1,2,3 etc. machine can make unwanted comparison.for example here we don't need to make comparison between the countries.So it is done when there is no need of any comparisons or where orders(like small,medium,large) does not matters.
If a certain country is present in that column that categorical column will show 1 else 0.
LabelEncoder -> encodes values without bothering for order.This creates a binary column for each category

#Making Dummy Variable

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehotencoder = OneHotEncoder(categorical_features=[0])
```

```python
X = onehotencoder.fit_transform(X).toarray()
```

```python
X
```

For variable y we don't need to use OneHotEncoder instead we can use LabelEncoder because Y is the dependent variable, and machine learning model will know that it is a category and that there is no order between the two.

```python
labelencoder_y = LabelEncoder()
```

```python
y = labelencoder_y.fit_transform(y)
```

```python
y
```

**Splitting the Dataset into the Training Set and Test Set**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)#(X,y) is the whole array
```

```python
X_train
```

```python
print('Size of Train and Test Set\nX_train = ',len(X_train),'\nX_test = ',len(X_test),'\ny_train = ',len(y_train),'\ny_test = ',len(y_test))
```

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
```

```python
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```
